```python
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier,GradientBoostingRegressor
from sklearn.model_selection import cross_val_score
import pandas as pd
```

```python
train = pd.read_csv('./data-science-london-scikit-learn/train.csv')
train_label = pd.read_csv('./data-science-london-scikit-learn/trainLabels.csv')
train_Y = train_label['y']
print(train_Y.head())
train.head()
```

```python
# describe the data
train.isnull().sum() # check nan data
train.describe()
```

```python
#x_train, x_test, y_train, y_test = train_test_split(train, train_Y, test_size=0.2, random_state=4)
```

```python
from sklearn.preprocessing import MinMaxScaler
MMEncoder = MinMaxScaler()
train_X = MMEncoder.fit_transform(train)
```

```python
# 建立模型
clf = GradientBoostingClassifier()
print(cross_val_score(clf, train_X, train_Y, cv=5).mean())


# 訓練模型
clf.fit(train_X, train_Y)
```

```python
# 設定要訓練的超參數組合
n_estimators = [100, 200, 300, 400, 500]
max_depth = [1, 3, 5, 7]
param_grid = dict(n_estimators=n_estimators, max_depth=max_depth)

## 建立搜尋物件，放入模型及參數組合字典 (n_jobs=-1 會使用全部 cpu 平行運算)
grid_search = GridSearchCV(clf, param_grid, scoring="accuracy", n_jobs=-1, verbose=2)

# 開始搜尋最佳參數
grid_result = grid_search.fit(train_X, train_Y)

# 預設會跑 3-fold cross-validadtion，總共 20 種參數組合，總共要 train 60 次模型
```

```python
# 印出最佳結果與最佳參數
print("Best Accuracy: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
# 使用最佳參數重新建立模型
clf_bestparam = GradientBoostingClassifier(max_depth=grid_result.best_params_['max_depth'],
                                           n_estimators=grid_result.best_params_['n_estimators'])

print(cross_val_score(clf_bestparam, train_X, train_Y, cv=5).mean())

# 訓練模型
clf_bestparam.fit(train_X, train_Y)
```

```python
# 預測real測試集
test = pd.read_csv('./data-science-london-scikit-learn/test.csv')
test.head()
test_X = MMEncoder.transform(test)
y_pred = clf_bestparam.predict(test_X)
print(y_pred)
id = range(1, 9001)
prd_opt = pd.DataFrame({'Id':id, 'Solution':y_pred})
print(prd_opt.head())

prd_opt.to_csv('Result.csv')
```
