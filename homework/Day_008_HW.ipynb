```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 作業
1. 請將 app_train 中的 CNT_CHILDREN 依照下列規則分為四組，並將其結果在原本的 dataframe 命名為 CNT_CHILDREN_GROUP
    * 0 個小孩
    * 有 1 - 2 個小孩
    * 有 3 - 5 個小孩
    * 有超過 5 個小孩

2. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，列出各組的平均 AMT_INCOME_TOTAL，並繪製 baxplot
3. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，對 AMT_INCOME_TOTAL 計算 [Z 轉換](https://en.wikipedia.org/wiki/Standard_score) 後的分數

```python
#1

group_names = ['0','1', '2', '3']
cut_rule = [0,0.99,2,5,7]
print(app_train['CNT_CHILDREN'].value_counts())
app_train['CNT_CHILDREN_GROUP'] = pd.cut(app_train['CNT_CHILDREN'].values, cut_rule,labels=group_names, include_lowest=True)
print(app_train['CNT_CHILDREN_GROUP'].head())
print(app_train['CNT_CHILDREN_GROUP'].value_counts())
app_train.loc[app_train['CNT_CHILDREN_GROUP'].isnull(),'CNT_CHILDREN_GROUP'] = '3'
print(app_train['CNT_CHILDREN_GROUP'].value_counts())
```

```python
#2-1

grp = ['CNT_CHILDREN_GROUP','TARGET']

grouped_df = app_train.groupby(grp)['AMT_INCOME_TOTAL']
print(grouped_df)
mean = grouped_df.mean()
print(mean)
app_train.boxplot(column='AMT_INCOME_TOTAL',by=grp)
```

```python
#2-2
"""
Your code here
"""
plt_column = 'AMT_INCOME_TOTAL'
plt_by = grp

app_train.boxplot(column=plt_column, by = plt_by, showfliers = False, figsize=(10,10))
plt.suptitle('')
plt.show()
```

```python
#3

print(grouped_df.describe())
app_train['AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET'] = grouped_df.apply(lambda x: (x-x.mean()) / x.std())
app_train[['AMT_INCOME_TOTAL','CNT_CHILDREN_GROUP','TARGET','AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET']].head()
```
